```python
import dotenv
import os
from sqlalchemy import create_engine
import pandas as pd

dotenv.load_dotenv(override=True)
username=os.getenv("username")
password=os.getenv("password")
host=os.getenv("host")
dbname=os.getenv("dbname")

conn_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
db_engine = create_engine(conn_string)

query = """SELECT * FROM dimproduct"""
df = pd.read_sql(query, db_engine)
db_engine.dispose()
#df.shape
df
```

```python
"""Esercizio 2/3 
• In media quanto pesano i prodotti? 
• Quanto pesa il più leggero? 
• Quanto pesa il più pesante? 
• Quanti prodotti pesano più di 100 Kg? 
• Quanto costano in media i prodotti (colonna DealerPrice)? 
• Se prendiamo un quarto di tutti i prodotti, in modo che siano i più costosi, quale sarà il loro range di prezzo?"""

#In media quanto pesano i prodotti? 
peso = df['Weight'].mean()
print("Il peso medio è: ", peso)

#Quanto pesa il più leggero?
peso_minimo = df['Weight'].min()
print("Il peso minimo è: ", peso_minimo)

#Quanto pesa il più pesante?
peso_massimo = df['Weight'].max()
print("Il peso massimo è: ", peso_massimo)
```

```python
#Quanti prodotti pesano più di 100 Kg? 
prodotti_100 = df[df['Weight'] > 100].shape[0]
print(prodotti_100)

#oppure
#prodotti_100 = len(df[df['Weight'] > 100])
```

```python
#Quanto costano in media i prodotti (colonna DealerPrice)?
media = df['DealerPrice'].mean()
print("In media i prodotti costano: ", round(media, 2))
```

```python
#Se prendiamo un quarto di tutti i prodotti, in modo che siano i più costosi, quale sarà il loro range di prezzo?
df_ordinato = df.sort_values(by = 'DealerPrice', ascending = False)

quarto = int(len(df_ordinato["DealerPrice"])/4)
print(quarto)
```

```python
#Qual è il prezzo medio per i prodotti di colore blu?

colore_blu = df[df['Color'] == 'Blue']
prezzo_medio_blu = colore_blu['DealerPrice'].mean()
print(prezzo_medio_blu)

#oppure
#colore_medio_blu = df[df['Color'] == 'Blue']['DealerPrice'].mean()
#print(colore_medio_blu)
```

```python
#Qual è il prezzo medio per i prodotti di colore rosso o nero? 
prodotti_neri_rossi = df[(df['Color'] == 'Red') | (df['Color'] == 'Black')]           # | questo significa OR
media_prezzi_neri_rossi =prodotti_neri_rossi['DealerPrice'].mean()
print(media_prezzi_neri_rossi)

#colore_medio_rossonero = df[df['Color'].isin(['Red', 'Black'])]
#prezzo_medio_rossonero = colore_medio_rossonero['DealerPrice'].mean()
#print(colore_medio_rossonero)
```

```python
#Qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg? 
prodotti_taglia42_peso10 = df[(df['Size'] == '42') & (df['Weight']>10)]    # & significa AND
prezzo_max = prodotti_taglia42_peso10['DealerPrice'].max()
prezzo_max

#siccome la taglia è un 'object' è stato scritto come una stinga, mentre il peso è un float, perciò posso scrivere il numero
```
